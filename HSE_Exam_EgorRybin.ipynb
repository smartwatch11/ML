My machine learning task is predicting class of Mushroom (edible or poisonous)

```python
# Importing the required library for data processing, works with CSV file
import pandas as pd

# Reading the dataset
df = pd.read_csv('mushrooms.csv', error_bad_lines=False)
```

```python
#some libraries for building graphs
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots

#Determine the number of rows and columns for future charts, and also their names
rows = 8
cols = 3
columns = ['class', 'cap-shape', 'cap-surface', 'cap-color', 'bruises', 'odor',
       'gill-attachment', 'gill-spacing', 'gill-size', 'gill-color',
       'stalk-shape', 'stalk-root', 'stalk-surface-above-ring',
       'stalk-surface-below-ring', 'stalk-color-above-ring',
       'stalk-color-below-ring', 'veil-type', 'veil-color', 'ring-number',
       'ring-type', 'spore-print-color', 'population', 'habitat']
fig = make_subplots(rows=rows, cols=cols,subplot_titles=columns ) #Create an area for our charts

#For each property, we build a distribution histogram
for i, col in enumerate(columns):
    r = int(np.ceil(((i+1)/cols)))
    c = i%3+1
    fig.append_trace(go.Histogram(x=df[col]), row=r,col=c)

    fig.update_layout(
    showlegend=False,
    autosize=False,
    width=1300,
    height=1300,
    title={
    'text':'<b>Overview of Distribution of Columns</b>',
         'y':0.95, 
         'x':0.5,
         'xanchor': 'center',
         'yanchor': 'top' 
    })
fig.show()
```

We see that some features do not occur at all, for example, veil-type is only partial and some other features are also not represented in this dataset. 

Let's see on the distribution of features for target column "class"

```python
#build another area for our histograms relative to the target class
rows = 8
cols = 3
columns = ['cap-shape', 'cap-surface', 'cap-color', 'bruises', 'odor',
       'gill-attachment', 'gill-spacing', 'gill-size', 'gill-color',
       'stalk-shape', 'stalk-root', 'stalk-surface-above-ring',
       'stalk-surface-below-ring', 'stalk-color-above-ring',
       'stalk-color-below-ring', 'veil-type', 'veil-color', 'ring-number',
       'ring-type', 'spore-print-color', 'population', 'habitat']
fig = make_subplots(rows=rows, cols=cols,subplot_titles=columns )

for i, col in enumerate(columns):
    r = int(np.ceil(((i+1)/cols)))
    c = i%3+1
    p = df[df['class']=='p']
    e = df[df['class']=='e']
    fig.append_trace(go.Histogram(x=p[col],marker_color = '#60391d',name='Poisonous'), row=r,col=c)
    fig.append_trace(go.Histogram(x=e[col],marker_color = '#ff9000',name='Edible'), row=r,col=c)

    
    fig.update_layout(
    showlegend=False,
    autosize=False,
    width=1300,
    height=1700,
    title={
    'text':'<b>Overview of Distribution of Columns for Poisonious & Edible Mushrooms</b>',
         'y':0.98, 
         'x':0.5,
         'xanchor': 'center',
         'yanchor': 'top' 
    })
fig.show()
```

From here we see that some features absolutely accurately indicate the edible type of mushroom, for example, the absence of odor indicates the edibility of the fungus.

```python
#create a separate array with the target class and clearly label the classes
#and then remove it from the main dataset and create a separate column for each feature
y = df["class"].map({"p":"poisonous","e":"edible"})
X = df.drop(["class"], axis=1)
X = pd.get_dummies(X, X.columns)
X.head()
```

```python
y
```

```python
#Using model_selection for splitting the dataset for train and test parts in a ratio of 70 to 30
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

The random forest algorithm was chosen as a model, since it is a very universal algorithm, and when adjusting hyperparameters, a high degree of learning can be achieved.

```python
#import from sklearn library: RandomForestClassifier - our model and GridSearchCV - for automatic selection of hyperparameters for models
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
model = RandomForestClassifier() #create the model
#Setting hyperparameters 
parameters = {
    'n_estimators': [50, 100, 150, 200],  #In our case - the number of trees - the more trees, the better the quality, but the RF setup and operation time also increases proportionally.
    'max_depth': [3, 5, 7, 10, 12, 15],   #Tree depth
    'min_samples_split': [5, 10, 15],     #The minimum number of objects at which splitting is performed
    'min_samples_leaf': [5, 10, 15]       #Limit on the number of objects in leaves
}

rfc = GridSearchCV(model, parameters, refit=True, cv=10)
# Train the model
rfc.fit(X_train, y_train)
# Run the trained model on the test dataset
y_pred = rfc.predict(X_test)
```

```python
#for i in range(len(y_pred)):
  #print(y_test.iat[i], y_pred[i])

#import metrics for Evaluating the trained model
from sklearn import metrics
print("Random Forest Classifier model accuracy:", metrics.accuracy_score(y_test, y_pred)*100) #Compare test data with predictions and print an accuracy
```

Accurancy of the model = 99.79. The model was well trained.
